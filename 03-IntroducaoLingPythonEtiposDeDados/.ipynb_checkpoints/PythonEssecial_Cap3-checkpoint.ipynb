Python Essencial CDN 

```python
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))
nota3 = float(input('Digite a terceira notaa: '))

media = (nota1 + nota2 + nota3) / 3
print(f'Média: {media:.1f}')

if media >= 6:
    print('Aprovado') 

elif media >= 4:
    print('Recuperação')

else: 
    print('Reprovado')
```

```python
# Calculo da media da natas de CDN Semestre
P1 = float(input('Digite a nota P1: ')) 
P2 = float(input('Digite a nota P2: '))
Trabalho = float(input('Digite a nota trabalho: ')) 
PI2 = float(input('Digite a nota PI2: ')) 

media = (P1* 0.35) + (P2 * 0.35) + (Trabalho * 0.10) + (PI2 * 0.20)
print(f'\nMédia: {media:.2f}')

if media >= 6: 
    print('Aprovado')
    
elif media <= 6: 
    print('Substutiva') 

else: 
    print('DP')
```

```python
# Calculo da media da natas de CDN Semestre
while True:
    nome = input('\nNome Aluno: ')
    
    P1 = float(input('Digite a nota P1: ')) 
    P2 = float(input('Digite a nota P2: '))
    Trabalho = float(input('Digite a nota trabalho: ')) 
    PI2 = float(input('Digite a nota PI2: ')) 
    
    media = (P1* 0.35) + (P2 * 0.35) + (Trabalho * 0.10) + (PI2 * 0.20)
    print(f'\nMédia: {media:.2f}')
    print(f'\nAluno: {nome}')
    
    if media >= 6: 
        print('Aprovado')
        
    elif media <= 6: 
        print('Substutiva') 
    
    else: 
        print('DP')
    
    continuar = input('\nDeseja calcular nota de outro aluno? (S/N)') 
    
    if continuar.upper( ) != 'S':
        break
```
