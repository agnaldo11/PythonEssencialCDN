## Comandos de Repetição

#### while

```python
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

```python
numeros = [1, 2, 3, 4, 5]
while numeros:             # O laço continua enquanto a lista não estiver vazia
    print(numeros.pop())   # Remove e imprime o último elemento da lista
```

```python
a = ['x', 'x', 'x', 'x', 'x']
while 'x' in a:
    print(a)
    a.pop()
```

#### for

```python
for letra in "Python":
    print(letra)
```

```python
frutas = ["maçã", "banana", "cereja", "laranja"]
for fruta in frutas:
    print(fruta)
```

#### A função range()

```python
for i in range(5):
    print(i)
```

```python
for i in range(2, 6):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(10, 0, -2):
    print(i)
```

#### break

```python
animais = ["gato", "cachorro", "papagaio", "tigre", "elefante"]
for animal in animais:
    if animal == "tigre":
        break
    print(animal)
```

#### continue

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    if numero == 3:
        continue
    print(numero)
```

#### Um exemplo para Finalizar: Tabuada!!

```python
# tabuada do 7
numero = 7
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```
