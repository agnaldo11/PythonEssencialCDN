## Comandos Condicionais

#### if

```python
if 5 > 3:
    print("Sim")
```

```python
if 10 > 5:
    print("10 é maior que 5")
    print("Essa linha também faz parte do bloco de código")
print("Essa linha está fora do bloco de código")
```

```python
if 42 > 99:
    print("Sim")  # Isso não será impresso porque a condição é falsa.
```

#### if - else

```python
if 42 > 99:
    print("Sim")
else:
    print("Não")
```

#### if - elif - else

```python
temperatura = 25

if temperatura > 30:
    print("Está muito quente!")
elif temperatura > 20:
    print("O clima está agradável.")
elif temperatura > 10:
    print("Está um pouco frio.")
else:
    print("Está muito frio!")
```

#### Um exemplo final...

```python
pontuacao = 72

if pontuacao >= 90:
    nota = 'A'
elif pontuacao >= 80:
    nota = 'B'
elif pontuacao >= 70:
    nota = 'C'  # Saída: C
elif pontuacao >= 60:
    nota = 'D'
else:
    nota = 'F'

print(f"Para a pontuação de {pontuacao}, a nota é {nota}")
```
