Os comandos condicionais permitem controlar o fluxo do programa com base em condições específicas. Em Python, os principais comandos condicionais são `if`, `elif` e `else`.

## Condicional `if`

A instrução `if` avalia uma condição e executa um bloco de código se a condição for verdadeira.

```python
if 5 == 5:
    print("Python para Ciência de Dados!")
```

```python
if True:
    print('Tudo está funcionando bem!')
```

```python
# Cuidado com a sintaxe
if 4 > 3
    print("Será que funciona!")
```

```python
# Cuidado com a sintaxe 2
if 4 > 3:
print("Será que funciona!")
```

```python
# Exemplo simples de uso de if com variável.
idade = 20

if idade >= 18:
    print("Você é maior de idade.")
```

## Condicional `if` com `else`

A instrução `else` complementa a `if` e é executada se a condição da `if` for falsa.

```python
# Exemplo de if com else
nota = 65

if nota >= 60:
    print("Você passou!")
else:
    print("Você reprovou.")
```

## Condicional `if` com `elif` e `else`

O `elif` permite adicionar condições adicionais entre o `if` e o `else`.

```python
# Exemplo de if com elif e else
pontuacao = 85

if pontuacao >= 90:
    print("Nota: A")
elif pontuacao >= 80:
    print("Nota: B")
elif pontuacao >= 70:
    print("Nota: C")
elif pontuacao >= 60:
    print("Nota: D")
else:
    print("Nota: F")
```

## Condicionais Aninhadas

Você pode aninhar instruções `if` dentro de outras instruções `if`.

```python
# Exemplo de condicionais aninhadas
temperatura = 30
umidade = 75

if temperatura > 25:
    if umidade > 70:
        print("Está quente e úmido.")
    else:
        print("Está quente, mas seco.")
else:
    print("Não está tão quente.")
```

## Condicionais com Operadores Lógicos

![Operadores Lógicos](https://piva.pro.br/pythonEssencial/operadores.jpg)

Os operadores lógicos `and`, `or` e `not` podem combinar múltiplas condições.

```python
# Exemplo com operadores lógicos
idade = 16
tem_permissao = True

if idade >= 18 or tem_permissao:
    print("Você pode dirigir.")
else:
    print("Você não pode dirigir.")
```

### Exemplo 1: Avaliação de Idade

```python
# Avaliando se uma pessoa é adolescente ou adulta
idade = 15

if idade < 13:
    print("Criança")
elif idade >= 13 and idade < 18:
    print("Adolescente")
else:
    print("Adulto")
```

### Exemplo 2: Avaliação de Notas com Input do Usuário

```python
# Avaliando a nota do aluno com base no input
nota = float(input("Digite a sua nota: "))

if nota >= 90:
    print("Excelente!")
elif nota >= 75:
    print("Bom!")
elif nota >= 50:
    print("Satisfatório.")
else:
    print("Precisa melhorar.")
```

3. Faça um Programa que peça os 3 lados de um triângulo. O programa deverá informar se os valores podem ser um triângulo. Indique, caso os lados formem um triângulo, se o mesmo é: equilátero, isósceles ou escaleno.
Dicas:
- Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro;
- Triângulo Equilátero: três lados iguais;
- Triângulo Isósceles: quaisquer dois lados iguais;
- Triângulo Escaleno: três lados diferentes;

```python
# solicita os valores de cada um dos lados
a = float(input("Digite o valor do lado 'a': "))
b = float(input("Digite o valor do lado 'b': "))
c = float(input("Digite o valor do lado 'c': "))

# verifica se os lados podem formar um triângulo
ehtriangulo = False
if (a+b > c) and (a+c > b) and (b+c > a):
    ehtriangulo = True

# se não formarem, exibir a mensagem adequada
if not ehtriangulo:
    print("Os lados não podem formar um triângulo!")
    
# se formarem um triângulo, verificar qual o tipo
if ehtriangulo:
    if (a == b == c):   # Equilátero
        print("É um triângulo Equilátero!")
    elif (a == b) or (b == c) or (a == c):  # Isósceles
        print("É um triângulo Isósceles!")
    else:   # Escaleno
        print("É um triângulo Escaleno!")
```
