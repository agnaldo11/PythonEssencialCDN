## **Estrutura de Dados Integrada**
As estruturas de dados integradas em Python permitem organizar e armazenar dados de forma eficiente. Existem principalmente 4 estruturas de dados integradas em Python: Listas (list), Tuplas (tuple), Conjuntos (set) e Dicionários (dict)

## Conjunto e seus Métodos

```python
set1 = {1, 2, 3, 4, 5}
set2 = {'a', 'b', 'c', 'd', 'e'}
print(set1)
print(set2)
```

#### add()

```python
set2.add('f')
print(set2)
```

#### remove()

```python
set2.remove('d')
print(set2)
```

#### update()

```python
# Conjunto 1 de utensílios domésticos
utensils_set1 = {"spoon", "fork", "knife", "plate", "bowl"}

# Conjunto 2 de utensílios domésticos
utensils_set2 = {"cup", "mug", "pan", "pot", "whisk"}

# Unindo os conjuntos em set1
utensils_set1.update(utensils_set2)
print(utensils_set1)
```

## Dicionários e seus Métodos

```python
# Exemplo de dicionário
student_info = {
    "name": "Alice",
    "age": 20,
    "course": "Data Science",
    "grade": "A"
}
print(student_info)
```

```python
# Acessando os valores usando as chaves
print(student_info["name"])
print(student_info["age"])
```

#### get()

```python
# Acessando os valores usando o método get()
print(student_info.get("name"))
print(student_info.get("age"))
```

```python
# Tentando acessar uma chave inexistente
print(student_info.get("address"))
```

```python
# Usando um valor padrão quando a chave não existe
print(student_info.get("address", "Address not available"))
```

#### keys()

```python
print(student_info.keys())
```

#### values()

```python
print(student_info.values())
```

#### items()

```python
print(student_info.items())
```

#### Alterando uma informação no dicionário

```python
print(student_info)
```

```python
student_info["name"] = "Silva"
print(student_info)
```

#### pop()

```python
print(student_info)
```

```python
student_info.pop("age")
print(student_info)
```

#### clear()

```python
print(student_info)
```

```python
student_info.clear()
print(student_info)
```

## Manipulação de Strings

```python
# Criando uma string
fruit = "banana"

# Acessando caracteres da string
print(fruit[0])
print(fruit[1])
print(fruit[2])
```

#### Principais Métodos de Strings...

#### upper()

```python
phrase = "I love programming"
print(phrase.upper())
```

#### lower()

```python
phrase = "I LOVE PROGRAMMING"
print(phrase.lower())
```

#### replace()

```python
message = "Hello World!"
new_message = message.replace("World", "Python")
print(new_message)
```

#### split()

```python
text = "Python is Fun"
words = text.split(" ")
print(words)
```

#### Strings Formatadas

```python
name = "Maria"
age = 21
message = f"Meu nome é {name} e eu tenho {age} anos."
print(message)
```
