## Módulos ou Pacotes em Python

#### Usando o módulo math

```python
import math

# Usando a função sqrt() do módulo math para calcular a raiz quadrada
numero = 16
raiz_quadrada = math.sqrt(numero)
print(f"A raiz quadrada de {numero} é {raiz_quadrada}")
```

#### Usando o Pacote numpy

```python
import numpy as np

# Criando um array simples com numpy
array = np.array([1, 2, 3, 4, 5])
print("Array criado com numpy:", array)
```

#### Importando funções específicas

```python
from math import pi, sin

# Usando as funções pi e sin diretamente
print(f"O valor de pi é {pi}")
print(f"O seno de 90 graus (em radianos) é {sin(pi/2)}")
```

```python
import matplotlib.pyplot as plt

# Dados para o gráfico
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
vendas = [150, 200, 250, 300, 350, 400]

# Criando o gráfico de linha
plt.plot(meses, vendas)
plt.title('Vendas nos Primeiros Seis Meses')
plt.xlabel('Meses')
plt.ylabel('Vendas (em unidades)')
plt.grid(True)
plt.show()
```
