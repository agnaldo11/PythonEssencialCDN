## Introdução à Ciência de Dados

#### 1. Carregar os Dados

```python
# Importando o pacote Pandas
import pandas as pd

# Carregar os dados
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

#### 2. Limpeza e Tratamento de Dados

```python
# Verificar as colunas disponíveis no DataFrame
print(df.columns)
```

```python
# Remover colunas irrelevantes
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Preencher valores faltantes, reatribuindo diretamente as colunas
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

#### 3. Análise Exploratória de Dados (EDA)

```python
# Importando os pacotes e bibliotecas para exibição de gráficos
import seaborn as sns
import matplotlib.pyplot as plt

# Distribuição das idades dos passageiros
sns.histplot(df['Age'], kde=False, bins=30)
plt.title("Distribuição das Idades dos Passageiros")
plt.show()
```

#### 4. Modelagem de Dados

```python
# Importando os pacotes e bibliotecas para criação do modelo
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Converter variáveis categóricas em numéricas
df = pd.get_dummies(df, drop_first=True)

# Separar variáveis independentes e dependentes
X = df.drop('Survived', axis=1)
y = df['Survived']

# Dividir em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinar o modelo
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

#### 5. Avaliação do Modelo

```python
# Avaliar a acurácia
accuracy = model.score(X_test, y_test)
print(f"Acurácia do Modelo: {accuracy * 100:.2f}%")
```
