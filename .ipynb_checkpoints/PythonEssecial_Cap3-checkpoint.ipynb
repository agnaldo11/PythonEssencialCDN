```python
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))
media = (nota1 + nota2)/2 
print(f'soma das notas: {media}')

if media >= 6:
    print('Aproivado') 

elif media >= 4:
    print('Recuperação')

else: 
    print('Reprovado')
```
