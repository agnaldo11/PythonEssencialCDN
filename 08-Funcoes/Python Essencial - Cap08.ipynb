## Funções

#### Uma função simples...

```python
def boas_vindas():
    print("Bem-vindo! Estamos felizes em ter você aqui.")
```

```python
boas_vindas()
```

```python
# Definindo uma função simples
def minha_funcao():
    print("Olá! Esta é uma função simples.")
```

```python
minha_funcao()
```

```python
# Definindo a função
def saudacao():
    print("Oi! Como você está?")
```

#### Uma função com parâmetro

```python
def saudacao_personalizada(nome):
    print(f"Olá, {nome}! Seja bem-vindo!")
```

```python
saudacao_personalizada("João")
```

```python
saudacao_personalizada("Maria")
```

#### Uma função com parâmetro e retorno de valor

```python
def soma(a, b):
    return a + b
```

```python
resultado = soma(3, 5)
print(f"A soma é: {resultado}")
```

#### Uma função que automatiza uma tarefa

```python
def calcular_desconto(preco, desconto):
    preco_com_desconto = preco - (preco * desconto / 100)
    return preco_com_desconto
```

```python
produto1 = calcular_desconto(100, 10)  # Calcula 10% de desconto em R$ 100
produto2 = calcular_desconto(200, 15)  # Calcula 15% de desconto em R$ 200

print(f"Preço do produto 1 com desconto: R$ {produto1}")
print(f"Preço do produto 2 com desconto: R$ {produto2}")
```

#### Função input() - Interação com o usuário

```python
# Pedindo o nome do usuário
nome = input("Por favor, digite seu nome: ")
print(f"Olá, {nome}! Seja bem-vindo ao nosso sistema.")
```

```python
print("Vamos somar dois números!")
num1 = input("Digite o primeiro número: ")
num2 = input("Digite o segundo número: ")
print(num1 + num2)
```

```python
print("Vamos somar dois números!")
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
print(f"A soma é: {num1 + num2}")
```
