# Parte 8 - Funções em Python

Este notebook aborda o conceito de funções em Python, sua definição, uso e diferentes formas de utilizá-las.

## Introdução às Funções

Funções são blocos de código que realizam uma tarefa específica e podem ser reutilizadas em diferentes partes de um programa.
Elas ajudam a modularizar o código, facilitando a manutenção e melhorando a clareza.

```python
# Exemplo de função simples:
def saudacao():
    print("Olá, Mundo!")
```

```python
# Chamando a função:
saudacao()
```

## Definindo Funções

A definição de uma função em Python começa com a palavra-chave `def`, seguida pelo nome da função e parênteses ().
O bloco de código que compõe a função é indentado.

```python
# Exemplo de uma função que imprime uma mensagem:
def exibir_mensagem():
    print("Esta é uma mensagem de uma função.")
```

```python
# Chamando a função:
exibir_mensagem()
```

## Argumentos e Parâmetros

Funções podem aceitar entradas chamadas de argumentos ou parâmetros, que permitem que o comportamento da função seja personalizado.

```python
# Exemplo com argumentos posicionais:
def saudar(nome, saudacao):
    print(f"{saudacao}, {nome}!")
```

```python
# Chamando a função com argumentos:
saudar("Carlos", "Olá")
```

```python
# Exemplo com argumentos nomeados:
saudar(nome="Ana", saudacao="Bom dia")
```

```python
saudar(saudacao="Boa Tarde", nome="Piva")
```

```python
# Usando valores padrão para parâmetros:
def apresentar(nome, idade=30):
    print(f"Nome: {nome}, Idade: {idade}")
```

```python
# Chamando a função sem o argumento opcional:
apresentar("João")
```

```python
# Chamando a função com o argumento opcional:
apresentar("Maria", 25)
```

## Retornando Valores

Funções podem retornar valores usando a palavra-chave `return`.
O valor retornado pode ser usado em outras partes do programa.

```python
# Exemplo de função que retorna a soma de dois números:
def somar(a, b):
    return a + b
```

```python
# Chamando a função e armazenando o resultado:
resultado = somar(5, 7)
print(f"Soma: {resultado}")
```

```python
# Funções sem `return` retornam `None` por padrão:
def funcao_sem_retorno():
    print("Esta função não retorna nada.")
```

```python
retorno = funcao_sem_retorno()
print(f"Retorno: {retorno}")
```

## Escopo de Variáveis

O escopo de uma variável refere-se à parte do programa onde a variável é reconhecida.
As variáveis podem ter escopo local ou global.

```python
# Exemplo de variáveis locais:
def calcular_area(largura, altura):
    area = largura * altura  # `area` é uma variável local
    return area

print(area)
```

```python
# Exemplo de variáveis globais:
global_var = "Eu sou global"

def mostrar_global():
    global_var = "Redefinida!!"
    print(global_var)  # Pode acessar a variável global

mostrar_global()
```

## Funções com Vários Argumentos

Python permite funções que aceitam um número variável de argumentos usando `*args` e `**kwargs`.

```python
# `*args` para argumentos posicionais variáveis:
def somar_varios(*args):
    soma = sum(args)
    return soma
```

```python
# Chamando a função com diferentes números de argumentos:
print(somar_varios(1, 2, 3)) 
```

```python
# Chamando a função com diferentes números de argumentos:
print(somar_varios(4, 5, 6, 7))
```

```python
# `**kwargs` para argumentos nomeados variáveis:
def mostrar_informacoes(**kwargs):
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")
```

```python
# Chamando a função com diferentes conjuntos de argumentos:
mostrar_informacoes(nome="Lucas", idade=32)
```

```python
# Chamando a função com diferentes conjuntos de argumentos:
mostrar_informacoes(produto="Notebook", preco=1200.00, quantidade=5)
```

## Funções Lambda

Funções lambda são funções anônimas de uma única linha definidas usando a palavra-chave `lambda`.
Elas são frequentemente usadas para operações simples que são passadas como argumentos para outras funções.

```python
# Exemplo de função lambda que calcula o quadrado de um número:
quadrado = lambda x: x ** 2
```

```python
# Chamando a função lambda:
print(quadrado(5)) 
```

```python
# Usando funções lambda com `sorted` para ordenar uma lista de tuplas por um critério específico:
alunos = [("João", 85), ("Ana", 78), ("Carlos", 92)]
```

```python
alunos_ordenados = sorted(alunos)
```

```python
print(alunos_ordenados) 
```

```python
alunos_ordenados = sorted(alunos, key=lambda aluno: aluno[1])
```

```python
print(alunos_ordenados) 
```

## Documentando Funções

Docstrings são strings de documentação que descrevem o propósito e o comportamento de uma função.
Elas são colocadas logo após a definição da função e podem ser acessadas pelo atributo `__doc__`.

```python
def calcular_fatorial(n):
    """
    Calcula o fatorial de um número.

    Parâmetros:
    n (int): O número para calcular o fatorial.

    Retorna:
    int: O fatorial de `n`.
    """
    if n == 0:
        return 1
    else:
        return n * calcular_fatorial(n - 1)
```

```python
# Exibindo a docstring da função:
print(calcular_fatorial.__doc__)
```

## Exercícios 

1. Faça um programa com uma função chamada somaImposto. A função possui dois parâmetros formais: taxaImposto, que é a quantia de imposto sobre vendas expressa em porcentagem e custo, que é o custo de um item antes do imposto. A função “altera” o valor de custo para incluir o imposto sobre vendas.

```python
def somaImposto(taxaImposto, custo):
    custo = custo + (custo * taxaImposto / 100.0)
    return custo
```

```python
taxa = float(input('Informe o valor da taxa de imposto: '))
custo = float(input('Informe o custo do produto: '))
custo = somaImposto(taxa, custo)

print(f'O preço com impostos é: {custo:.2f}')
```

2. Faça um programa que use a função valorPagamento para determinar o valor a ser pago por uma prestação de uma conta. O programa deverá solicitar ao usuário o valor da prestação e o número de dias em atraso e passar estes valores para a função valorPagamento, que calculará o valor a ser pago e devolverá este valor ao programa que a chamou. O programa deverá então exibir o valor a ser pago na tela. Após a execução o programa deverá voltar a pedir outro valor de prestação e assim continuar até que seja informado um valor igual a zero para a prestação. Neste momento o programa deverá ser encerrado, exibindo o relatório do dia, que conterá a quantidade e o valor total de prestações pagas no dia. O cálculo do valor a ser pago é feito da seguinte forma. Para pagamentos sem atraso, cobrar o valor da prestação. Quando houver atraso, cobrar 3% de multa, mais 0,1% de juros por dia de atraso

```python
def valorPagamento(valor, diasAtraso):
    if (valor < 0):
        return None
    if (diasAtraso > 0):
        multa = valor * 0.03
        adicionalAtraso = valor * (diasAtraso * 0.01)
        return valor + multa + adicionalAtraso
    else:
        return valor
```

```python
# Entrada de Dados
valor = 1
while (valor != 0):
    valor = float(input('Informe o valor da prestação: '))
    if (valor != 0):
        diasAtraso = int(input('Informe a qtd de dias de atraso: '))
        print(f"Valor a ser pago: {valorPagamento(valor, diasAtraso):.2f}")
```
